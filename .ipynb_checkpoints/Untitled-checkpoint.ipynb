```python
import pandas as pd
import numpy as np
import seaborn as sn
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()

from sklearn.model_selection import KFold, train_test_split, cross_val_score
from sklearn import metrics
from sklearn.metrics import classification_report ,confusion_matrix

from numpy import mean


veriler = pd.read_csv('ObesityDataSet_raw_and_data_sinthetic.csv')
#print(veriler)

gender=veriler.iloc[:,0:1].values
le=LabelEncoder() #verileri 0 ile 1 yapıyıor kadın=1,erkek=0,sadece 1 yada 0 olarak gösteriyior
gender[:,0]=le.fit_transform(gender[:,0])
#print(gender)

#encoder verileri dönüştürmeyi sağlıyıor
ohe=OneHotEncoder()  #0 ile 1 olan verileri kadın ise=> 1 0 erkek ise =>0 1 olarak gösteriyior
gender=ohe.fit_transform(gender).toarray()
#print(gender)

cinsiyet=pd.DataFrame(data=gender[:,1],index=range(2111),columns=['Male'])#sadece erkek olup olmadıgını aldım çünkü kadın olup
#olmadığıda eklenince modeli fazla etkileyebilir oyuzden sadece erkek olup olmamasına bakılıyıor
#print(cinsiyet)

age=veriler.iloc[:,1].values
age=pd.DataFrame(data=age,index=range(2111),columns=['Age'])

height=veriler.iloc[:,2].values
height=pd.DataFrame(data=height,index=range(2111),columns=['Height'])

weight=veriler.iloc[:,3].values
weight=pd.DataFrame(data=weight,index=range(2111),columns=['Weight'])

family_history_with_overweight=veriler.iloc[:,4:5].values  #aile üyelerinden biri aşırı kilolumu 
family_history_with_overweight[:,0]=le.fit_transform(family_history_with_overweight[:,0])  
family_history_with_overweight=ohe.fit_transform(family_history_with_overweight).toarray()
family_history_with_overweight=pd.DataFrame(data=family_history_with_overweight[:,1],index=range(2111),columns=['family_history_with_overweight_Yes'])
#sadece ailede aşırı kilolu biri olmasına göre  0-1 olacak 
#print(family_history_with_overweight)

FAVC  =veriler.iloc[:,5:6].values  #Sık sık yüksek kalorili yiyeceklerin yenilip yenilmemesi
FAVC  [:,0]=le.fit_transform(FAVC [:,0])  
FAVC  =ohe.fit_transform(FAVC).toarray()
FAVC  =pd.DataFrame(data=FAVC[:,1] ,index=range(2111),columns=['YuksekKaloriEvet'])#Sadece yüksekkolarili yiyip yemediği alındı
#print(FAVC)

FCVC =veriler.iloc[:,6:7].values  #öğünlerde sebzeli yemek yenip yenmediği
FCVC =pd.DataFrame(data=FCVC,index=range(2111),columns=['FCVC'])#1,2,3 never,sometimes,always
#print(FCVC)

NCP=veriler.iloc[:,7:8].values  #günde kaç ana öğün yemek yendiği
NCP =pd.DataFrame(data=NCP,index=range(2111),columns=['NCP'])#1,2,3   1-2arası , 3 , 3 ten fazla

CAEC=veriler.iloc[:,8:9].values  #öğün arasında yemek yenip yenmediği
CAEC[:,0]=le.fit_transform(CAEC[:,0])  
CAEC=ohe.fit_transform(CAEC).toarray()
CAEC=pd.DataFrame(data=CAEC,index=range(2111),columns=['Always','Frequently','Sometimes','No'])
#print(CAEC.head(17))

SMOKE=veriler.iloc[:,9:10].values  #sigara içip içmediği
SMOKE[:,0]=le.fit_transform(SMOKE[:,0])  
SMOKE=ohe.fit_transform(SMOKE).toarray()
SMOKE=pd.DataFrame(data=SMOKE[:,1],index=range(2111),columns=['SMOKE_Yes'])#sigara içtiğinde 1 alacak
#print(SMOKE)

CH2O=veriler.iloc[:,10:11].values  #vucuttaki organik bir birleşik
CH2O=pd.DataFrame(data=CH2O,index=range(2111),columns=['CH2O'])
#print(CH2O)

SCC=veriler.iloc[:,11:12].values  #günlük yenilen kalorilerin takip edilip edilmemesi etmiyiorsa 1 olacak
SCC[:,0]=le.fit_transform(SCC[:,0])  
SCC=ohe.fit_transform(SCC).toarray()
SCC=pd.DataFrame(data=SCC[:,0],index=range(2111),columns=['YedigiKaloriyiTakipEtmiyior'])
#print(SCC)

FAF=veriler.iloc[:,12:13].values  #
FAF=pd.DataFrame(data=FAF,index=range(2111),columns=['Yes'])

TUE=veriler.iloc[:,13:14].values  #
TUE=pd.DataFrame(data=TUE,index=range(2111),columns=['Yes'])

CALC=veriler.iloc[:,14:15].values  #
CALC[:,0]=le.fit_transform(CALC[:,0])  
CALC=ohe.fit_transform(CALC).toarray()
CALC=pd.DataFrame(data=CALC,index=range(2111),columns=['Always','Frequently','Sometimes','No'])
#print(CALC)

MTRANS=veriler.iloc[:,15:16].values  #Genellikle Kullanılan ulaşım aracı
MTRANS[:,0]=le.fit_transform(MTRANS[:,0])  
MTRANS=ohe.fit_transform(MTRANS).toarray()
MTRANS=pd.DataFrame(data=MTRANS,index=range(2111),columns=['Automobile','Bike','Motorbike','PublicTransportain','Walking'])
#print(MTRANS.head(12))

NObeyesdad=veriler.iloc[:,16:17].values  #ObeziteDurumu   InsufficientWeight=YetersizAğırlık
#NObeyesdad[:,0]=le.fit_transform(NObeyesdad[:,0])  
#NObeyesdad=ohe.fit_transform(NObeyesdad).toarray()
#NObeyesdad=pd.DataFrame(data=NObeyesdad,index=range(2111),columns=['InsufficientWeight','NormalWeight','ObesityType_1',
#'ObesityType_2','ObesityType_3','OverwightLevel_1','OverwightLevel_2'])
#print(NObeyesdad.head(200))

#dataframleri concat ile birleşir
a=pd.concat([cinsiyet,age],axis=1)
b=pd.concat([a,height],axis=1)
c=pd.concat([b,weight],axis=1)
d=pd.concat([c,family_history_with_overweight],axis=1)
e=pd.concat([d,FAVC],axis=1)
f=pd.concat([e,FCVC],axis=1)
g=pd.concat([f,NCP],axis=1)
h=pd.concat([g,CAEC],axis=1)
j=pd.concat([h,SMOKE],axis=1)
j1=pd.concat([j,CH2O],axis=1)
j2=pd.concat([j1,SCC],axis=1)
j3=pd.concat([j2,FAF],axis=1)
j4=pd.concat([j3,TUE],axis=1)
j5=pd.concat([j4,CALC],axis=1)
j6=pd.concat([j5,MTRANS],axis=1)
#j7=pd.concat([j6,NObeyesdad],axis=1)
#print(j7)

#LojistikRegresyon
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression(max_iter=2200,random_state=0)

x_train,x_test,y_train,y_test=train_test_split(j6,NObeyesdad.ravel(),test_size=0.25,random_state=0)
X_Train = sc.fit_transform(x_train)
X_Test =sc.fit_transform(x_test)
lr.fit(X_Train,y_train)
y_pred=lr.predict(X_Test)

cm=confusion_matrix(y_test,y_pred)
cm=pd.crosstab(y_test,y_pred)
sn.heatmap(cm, linewidths=1, annot=True,fmt='g',cbar=False)


cv = KFold(n_splits=10, random_state = 0, shuffle=True)

accuracy = cross_val_score(lr, X_Test, y_test, scoring='accuracy', cv=cv, n_jobs=-1)
f1_micro = cross_val_score(lr, X_Test, y_test, scoring='f1_weighted', cv=cv, n_jobs=-1)
recall = cross_val_score(lr, X_Test, y_test, scoring='recall_weighted', cv=cv, n_jobs=-1)
precision = cross_val_score(lr, X_Test, y_test, scoring='precision_weighted', cv=cv, n_jobs=-1)
roc_auc = cross_val_score(lr, X_Test, y_test, scoring='roc_auc_ovo_weighted', cv=cv, n_jobs=-1)

print('Ortalama Accuracy: %.5f' % (mean(accuracy)))
print('Ortalama f1_Score: %.5f' % (mean(f1_micro)))
print('Ortalama recall: %.5f' % (mean(recall)))
print('Ortalama precision: %.5f' % (mean(precision)))
print('Ortalama roc_auc: %.5f' % (mean(roc_auc)))
```

```python
#K-nn
from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier(n_neighbors=5, metric='minkowski')

classifier.fit(X_Train,y_train)
y_pred2=classifier.predict(X_Test)

cm=confusion_matrix(y_test,y_pred2)
cm=pd.crosstab(y_test,y_pred2)
sn.heatmap(cm, linewidths=1, annot=True,fmt='g',cbar=False)

accuracy = cross_val_score(classifier, X_Test, y_test, scoring='accuracy', cv=cv, n_jobs=-1)
f1_micro = cross_val_score(classifier, X_Test, y_test, scoring='f1_weighted', cv=cv, n_jobs=-1)
recall = cross_val_score(classifier, X_Test, y_test, scoring='recall_weighted', cv=cv, n_jobs=-1)
precision = cross_val_score(classifier, X_Test, y_test, scoring='precision_weighted', cv=cv, n_jobs=-1)
roc_auc = cross_val_score(classifier, X_Test, y_test, scoring='roc_auc_ovo_weighted', cv=cv, n_jobs=-1)

print('Ortalama Accuracy: %.5f' % (mean(accuracy)))
print('Ortalama f1_Score: %.5f' % (mean(f1_micro)))
print('Ortalama recall: %.5f' % (mean(recall)))
print('Ortalama precision: %.5f' % (mean(precision)))
print('Ortalama roc_auc: %.5f' % (mean(roc_auc)))
```

```python
#Naive Bayes
c_train,c_test,d_train,d_test=train_test_split(j6,NObeyesdad.ravel(),test_size=0.25,random_state=0)

C_train = sc.fit_transform(c_train)
C_test = sc.transform(c_test)
                        
from sklearn.naive_bayes import GaussianNB
GNB = GaussianNB()
GNB.fit(C_train, d_train)
c_pred = GNB.predict(C_test)

Ncm=confusion_matrix(d_test,c_pred)
Ncm=pd.crosstab(d_test,c_pred)
sn.heatmap(Ncm, linewidths=1, annot=True,fmt='g',cbar=False)

accuracy3 = cross_val_score(GNB, C_test, d_test, scoring='accuracy', cv=cv, n_jobs=-1)
f1_micro3 = cross_val_score(GNB, C_test, d_test, scoring='f1_weighted', cv=cv, n_jobs=-1)
recall3 = cross_val_score(GNB, C_test, d_test, scoring='recall_weighted', cv=cv, n_jobs=-1)
precision3 = cross_val_score(GNB, C_test, d_test, scoring='precision_weighted', cv=cv, n_jobs=-1)
roc_auc3 = cross_val_score(GNB, C_test, d_test, scoring='roc_auc_ovo_weighted', cv=cv, n_jobs=-1)

print('Ortalama Accuracy: %.5f' % (mean(accuracy3)))
print('Ortalama f1_Score: %.5f' % (mean(f1_micro3)))
print('Ortalama recall: %.5f' % (mean(recall3)))
print('Ortalama precision: %.5f' % (mean(precision3)))
print('Ortalama roc_auc: %.5f' % (mean(roc_auc3)))
```
